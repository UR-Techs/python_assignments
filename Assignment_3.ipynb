# Calculator

```python
val_1 = int(input("Enter the first value"))
val_2 = int(input("Enter the second value"))
operator = int(input(" Enter the number to select the respective operation: \n1. Add \n2. Subtract \n3. Multiply \n4. Divide \n5. Power\n"))
result = 0

if operator == 1:
    result = val_1 + val_2
    print("The sum is ",result)
elif operator == 2:
    result = val_1 - val_2
    print("The difference is ",result)
elif operator == 3:
    result = val_1 * val_2
    print("The product is ",result)
elif operator == 4:
    result = val_1 / val_2
    print("The quotient is ",result)
elif operator == 5:
    result = val_1 ** val_2
    print("The sum is ",result)
else:
    print("Please enter the correct number for the operation you want to perform.")
    
```

# Check Numeric Value In The List Using For Loop

```python
your_list = input("Enter the elements of your list and separate each item by ,")

is_numeric = 0

for item in your_list:
    if item.isdigit():
        print(item, "is a numeric value in your list")
        is_numeric += 1
    
if is_numeric == 0:
    print("There is no numeric value in your list")
    
```

# Sum All The Numeric Items In A Dictionary

```python
exam_scores = {
    "eng": 50,
    "urdu": 48,
    "math": 45,
    "isl": 45,
    "comp":50
}

sum = 0

for score in exam_scores.values():
    sum = sum + score

print("Your obtained marks are: ",sum)
```

# Identify Duplicate Values in a list

```python

your_list = input("Enter the elements of your list and separate each item by space").split()

item_dic = {}

for item in your_list:
    if item in item_dic.keys():
        item_dic[item] = item_dic[item] + 1
    else: 
        item_dic[item] = 1

for item in item_dic:
    if item_dic[item] > 1:
        print(f"{item}  has been repeated " + item_dic[item] + " times.")
print(item_dic)
```

# Check If a Key Already Exists in a Dictionary

```python
check_key = input("Enter a key to check: ")
basic_python = {
    "data_types": "done",
    "variables": "done",
    "input": "done",
    "output": "done",
    "file_handling": "pending"
               }
if check_key in basic_python.keys():
    print(check_key," exists in the dictionary")
else:
    print(check_key," doesn't exist in the dictionary")
    
```
